```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')

path = "/content/drive/MyDrive/weight of copper (1).csv"
df = pd.read_csv(path)
df
```

```python
plt.xlabel('POWER RATING')
plt.ylabel('AMOUNT OF COPPER')
plt.title('COPPER PREDICTION IN GENERATOR')
plt.hist(df[['power_rating', 'weight_of_copper']])
```

```python
from sklearn.model_selection import train_test_split
x=np.array(df['power_rating'])
y=df['weight_of_copper']
x_train,x_test,y_train,y_test=train_test_split(x.reshape(-1,1),y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```
